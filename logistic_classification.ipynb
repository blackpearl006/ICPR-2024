```python
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")

path = '/Users/ninad/Documents/_CBR/Peris-Dataset/CSV/TLSA50'
label_dict = {'HC': 0, 'MCI': 1}
random_state = 2
networks = ['DMN','CO','FP','OP','CB','SM']
accuracy_dict = {}

for descriptor_csv in ['h0.csv', 'h1.csv', 'h2.csv']:
    accuracy_dict[descriptor_csv] = {}
    for bnetwork in networks:
        X, y = [], []
        bnetwork_path = os.path.join(path, bnetwork)
        
        if os.path.isdir(bnetwork_path):
            for class_name in label_dict.keys():
                class_dir = os.path.join(bnetwork_path, class_name)
                
                if os.path.isdir(class_dir):
                    for file_name in os.listdir(class_dir):
                        if file_name.endswith(descriptor_csv):
                            file_path = os.path.join(class_dir, file_name)
                            correlation_matrix = pd.read_csv(file_path, index_col=0)
                            upper_triangular = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))
                            flattened_values = upper_triangular.stack().values
                            X.append(flattened_values)
                            y.append(label_dict[class_name])
        
        scaler = StandardScaler()
        X = scaler.fit_transform(X)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)

        lrclf = LogisticRegression(random_state=random_state)
        lrclf.fit(X_train, y_train)
        y_pred = lrclf.predict(X_test)
        
        cv_scores1 = cross_val_score(lrclf, X, y, cv=5)
        cv_scores2 = cross_val_score(lrclf, X, y, cv=10)
        
        accuracy_dict[descriptor_csv][bnetwork] = (round(np.mean(cv_scores1) * 100, 2), round(np.mean(cv_scores2) * 100, 2))

# Convert accuracy_dict to pandas DataFrame
accuracy_df = pd.DataFrame.from_dict(accuracy_dict, orient='index')
accuracy_df.T
```
